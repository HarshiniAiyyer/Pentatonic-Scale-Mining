### Here is the main crux of all the struggle I went through so far.

I found the median; and then i try to find the closest note frequency to the median. That will be our target variable for ML model we are going to implement.

```python
import numpy as np
import pandas as pd

# Define the mapping
mapping = {
    'D': [110, 220],
    'S': [130.81, 261.63],
    'R': [146.83, 293.66],
    'G': [164.81, 329.63],
    'P': [196]
}


# Define a function to find the closest note for each row
def find_closest_note(row):
    median = row['Median']
    closest_note = ''
    closest_distance = np.inf
    for note, values in mapping.items():
        for value in values:
            distance = abs(median - value)
            if distance < closest_distance:
                closest_note = note
                closest_distance = distance
    return closest_note

# Apply the function to the dataset and create a new column
copyres['closest_note'] = copyres.apply(find_closest_note, axis=1)

# Print the updated dataset
print(copyres)
```
